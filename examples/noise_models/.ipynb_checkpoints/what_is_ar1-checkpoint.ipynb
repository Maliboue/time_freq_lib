## AR(1) process:

$$
 x_k = \alpha x_{k-1} + \epsilon_k,
$$

where $k=0,\ldots,n-1$ is the index, $\epsilon_k$ is white noise.

The coefficient $\alpha$ is the slope between $x_k$ and $x_{k-1}$. Let's visualize this

```python
t = np.arange(0, T, 1/fs)
alpha=0.8
ps = dict(
    intercept = 0,
    ar_coeffs = [alpha],
    noise_std = 1,
)
s = cs.simulate_ar_surrogate(ps, n=t.size)

s1 = s[:-1]
s2 = s[1:] # delayed
```

```python
plt.scatter(s2,s1, label='$x_k$ vs $x_{k-1}$')
plt.plot([s2.min(), s2.max()], alpha*np.array([s2.min(), s2.max()]), color='k', label='Identity line')
plt.legend()
```
