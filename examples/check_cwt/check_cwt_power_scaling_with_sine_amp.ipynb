```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
import pywt

fs = 2.0
dt = 1 / fs
T = 2000
t = np.arange(0, T, dt)

f0 = 0.024
A = 1.0

x = A * np.sin(2 * np.pi * f0 * t)

print("Time-domain power:", np.mean(x**2))
```

```python
wavelet = "cmor8-0.5"

freqs = np.linspace(0.01, 0.04, 100)

# Convert target frequencies to scales
scales = pywt.frequency2scale(wavelet, freqs * dt)

coef, freqs_cwt = pywt.cwt(
    x,
    scales,
    wavelet,
    sampling_period=dt
)

power_cwt = np.abs(coef)**2
```

```python
res = []

for A in np.arange(0.2, 15, 0.2):
    x = A * np.sin(2 * np.pi * f0 * t)

    coef, _ = pywt.cwt(
        x,
        scales,
        wavelet,
        sampling_period=dt
    )

    power_cwt = np.abs(coef)**2
    cwt_band_power_raw = np.trapz(power_cwt, freqs_cwt, axis=0)

    print(A, cwt_band_power_raw.mean())
    
    res.append([A, cwt_band_power_raw.mean()])
    
# res = np.array(res)
```

```python
plt.plot(res[:, 0], res[:, 1]/res[:, 0]**2)
plt.ylabel('CWT Power / A$^2$')
plt.xlabel('Oscillation amplitude, A')
```
