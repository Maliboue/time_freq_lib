```python
import numpy as np
import matplotlib.pyplot as plt

# reproducibility
rng = np.random.default_rng(42)

# time axis
fs = 100                      # sampling frequency (Hz)
T = 20                        # duration (s)
t = np.arange(0, T, 1/fs)

# components
slow_drift = 0.05 * t + 0.5 * np.sin(2*np.pi*0.03*t)
oscillation = 1.0 * np.cos(2*np.pi*2*t)   # 2 Hz cosine
white_noise = rng.normal(size=len(t))

# combined signal
x = slow_drift + oscillation + white_noise

# plot
plt.figure(figsize=(12,4))
plt.plot(t, x, label='signal')
plt.plot(t, slow_drift, '--', lw=2, label='slow drift')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.tight_layout()
plt.show()
```
