## Beating cosine, no noise

```python
f1 = 2 # Hz
df = 0.7 # Hz
fs = 100
t,s = sg.cosine(10, f1, fs)
s = s  + sg.cosine(10, f1+df, fs)[1] + sg.cosine(10, f1+2*df, fs, phase=np.pi)[1] + sg.cosine(10, f1+3*df, fs, phase=np.pi/3, amplitude=1.5)[1]

_, (ax,ax1) = wf.plot_timeseries_and_its_cwt(s, t, f1=1, f2=7, method='fft', show_smoothed=False)
```

```python
plt.plot(*periodogram(s, fs=100))
plt.xlim(0.5,5)
plt.gca().set(xlabel='Frequency (Hz)', ylabel='Periodogram value (1/Hz)')
```
