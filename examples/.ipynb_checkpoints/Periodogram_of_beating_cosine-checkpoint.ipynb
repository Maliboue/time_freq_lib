# Periodogram of beating cosine when only quarter of beating period is in the sampled data

```python
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sys import path
path.append('../signals/')

import signals as sg
```

```python
from scipy.signal import periodogram
```

# No noise

## Duration is long enough to resolve two components

```python
dur = 100
t, x = sg.cosine(dur, 0.5, 100)
x = x + sg.cosine(dur, 0.52, 100, amplitude=0.5)[1]
```

```python
plt.plot(t,x)
```

```python
plt.step(*periodogram(x, fs=100), where='mid')
plt.xlim(0.3,0.7)
```

```python
(50+12)/dur
```

## Duration is too short to resolve the two components. They smear into one broad peak

```python
dur = 20
t, x = sg.cosine(dur, 0.5, 100)
x = x + sg.cosine(dur, 0.52, 100, amplitude=0.5)[1]
```

```python
plt.plot(t,x)
```

```python
plt.step(*periodogram(x, fs=100), where='mid')
plt.xlim(0.3,0.7)
```

```python
(14+1)/dur
```

# With noise

## Duration is long enough to resolve two components

```python
dur = 100
t, x = sg.cosine(dur, 0.5, 100)
x = x + sg.cosine(dur, 0.52, 100, amplitude=0.5)[1] + np.random.randn(t.size)*7
```

```python
plt.plot(t,x)
```

```python
plt.step(*periodogram(x, fs=100), where='mid')
plt.xlim(0.3,0.7)
```
