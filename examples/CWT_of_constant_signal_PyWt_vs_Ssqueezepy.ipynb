```python
import numpy as np
import matplotlib.pyplot as plt
from sys import path
path.append('../wavelet/')
import wavelet_funcs as wf
```

## Constant signal - CWT coefficients must be zero

```python
fs = 100
T = 10
t = np.arange(0, T, 1/fs)
s = np.ones(t.size)

_, (ax,ax1) = wf.plot_timeseries_and_its_cwt(s, t, f1=0.1, f2=5, method='fft', show_smoothed=False, precision=16)
```

```python
res = wf.cwt_with_coi(s, t, 0.1, 5, method='fft', precision=16)
```

```python
x = abs(res.wt_coefs).mean('t')/res.scales
```

```python
x.plot()
```

#### CWT coefficients are not zero for PyWavelets...
